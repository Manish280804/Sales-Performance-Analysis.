```python
import pandas as pd
import sqlalchemy
import seaborn as sns
import matplotlib.pyplot as plt

engine = sqlalchemy.create_engine("mysql+pymysql://root:qwerty@localhost/e_commerce")

df = pd.read_sql("SELECT * FROM Orders", engine)

df["OrderDate"] = pd.to_datetime(df["OrderDate"], dayfirst=True, errors="coerce")

print("Missing Values:\n", df.isnull().sum())

monthly_sales = df.groupby(df["OrderDate"].dt.to_period("M"))["Sales"].sum()
monthly_sales.plot(kind="line", title="Monthly Sales Trend", ylabel="Sales", xlabel="Month")
plt.savefig("monthly_sales.png")
plt.show()

sns.barplot(x="Category", y="Profit", data=df)
plt.title("Profit by Category")
plt.savefig("profit_by_category.png")
plt.show()

region_sales = df.groupby("Region")["Sales"].sum()
region_sales.plot(kind="pie", autopct="%1.1f%%", title="Sales by Region")
plt.savefig("sales_by_region.png")
plt.show()

top_products = df.groupby("ProductName")["Sales"].sum().nlargest(10)
top_products.plot(kind="bar", title="Top 10 Products by Sales")
plt.savefig("top_products.png")
plt.show()

print("Total Sales:", df["Sales"].sum())
print("Total Profit:", df["Profit"].sum())
print("Average Order Value:", df["Sales"].mean())
```
